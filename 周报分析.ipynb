## 财务数据 (g_caiwu.dwm_order_report_d)

```python
def fin_get_need_data(fin_whole,
                      need_wk_begin = 16,
                      need_wk_end   = 17,
                      need_city_id  = 1,
                      need_comm_product_id = 1,
                      need_facs = ['gmv',
                              'driver_cut',
                              'leasing_company_cut',
                              'driver_vat',
                              'p_adjust_fee',
                              'p_total_fee_closed',
                              'spam_rev',
                              'd_cancel_fee_divide',
                              'driver_cut_adjust',
                              'cancel_fee',
                              'payment_on_behalf_fee_closed',
                              'subsidy_b',
                              'subsidy_c',
                              'business_income',
                              'business_profit',
                              'finish_order_cnt',
                              'driver_vat_xx'
                              ]
                 ):
    p_df = pd.pivot_table(fin_whole,
                          index=['week','city_id','common_product_id'], 
                          values = need_facs, 
                          aggfunc=sum)

    query_str = "city_id=={}&\
                 common_product_id=={}&\
                 week>={}&\
                 week<={}".format(need_city_id, need_comm_product_id, need_wk_begin, need_wk_end) 
    f_df = p_df.query(query_str).reset_index()
    return f_df
```

```python
#财务数据_20190504
fin_whole = pd.read_csv('财务数据_20190504.csv').sort_values(['dt'])

fin_whole['week'] = fin_whole['dt'].apply(lambda x:dt.datetime.strptime(x, '%Y-%m-%d').isocalendar()[1])

comm_prod_dict = {1: [3,7,99,20,314],
                  2: [1,6],
                  3: [9]}
def get_common_prod_id(x):
    if x in comm_prod_dict[1]: return 1
    if x in comm_prod_dict[2]: return 2
    if x in comm_prod_dict[3]: return 3
    else:                      return -1
fin_whole['common_product_id'] = fin_whole['product_category'].apply(lambda x:get_common_prod_id(x))

fin_whole['driver_vat_xx'] = fin_whole['driver_vat'] * 0.12

fin_whole.head(3)
```

```python
f_df = fin_get_need_data(fin_whole, 16,17)
f_df
```

## 北极星

```python
def star_get_need_data(data,
                  need_dt_begin       = '2019-04-20',
                  need_dt_end         = '2019-05-03',
                  need_city_id        = 1,
                  need_pd             = 'fast',
                  need_product_id     = None ,
                  need_sub_product_id = -10000,
                  need_channel_id     = -10000,
                  need_private_dim_id = -10000
                 ):
    cond = 1
    if need_dt_begin and need_dt_end:
        cond_dt = (data['stat_date'].apply(lambda x:x>=need_dt_begin and x<=need_dt_end))
        cond = cond_dt
    if need_city_id:
        cond_city = (data['city_id']==need_city_id)
        cond = cond & cond_city
    if need_pd:
        cond_pd = (data['pd']==need_pd)
        cond = cond & cond_pd
    if need_product_id:
        cond_product_id = (data['product_id']==need_product_id)
        cond = cond & cond_product_id
    if need_sub_product_id:
        cond_sub_product_id = (data['sub_product_id']==need_sub_product_id)
        cond = cond & cond_sub_product_id
    if need_channel_id:
        cond_channel_id = (data['channel_id']==need_channel_id)
        cond = cond & cond_channel_id
    if need_private_dim_id:
        cond_private_dim_id = (data['product_private_dim_id']==need_private_dim_id)
        cond = cond & cond_private_dim_id

    new_data = data[cond]
    new_data = new_data.sort_values('stat_date').reindex()
    return new_data
```
